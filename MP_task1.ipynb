```python
from os.path import join
import sys
import matplotlib.pyplot as plt
import numpy as np


def load_data(load_dir, bid):
    SIZE = 512
    u = np.zeros((SIZE + 2, SIZE + 2))
    u[1:-1, 1:-1] = np.load(join(load_dir, f"{bid}_domain.npy"))
    interior_mask = np.load(join(load_dir, f"{bid}_interior.npy"))
    return u, interior_mask


load_dir = r"C:\Users\rikke\OneDrive - Danmarks Tekniske Universitet\Universitet\Kandidat - MMC\Python and High Performance Computing\MP_Data\modified_swiss_dwellings"

data_bids_avaible = ['25', '76', '108', '124']

for bid in data_bids_avaible:
    u, interior_mask = load_data(load_dir, bid)
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    im = axs[0].imshow(u, cmap='hot')
    axs[0].set_title(f'Building {bid} - Temperature Distribution')

    axs[1].imshow(interior_mask, cmap='gray')
    axs[1].set_title(f'Building {bid} - Interior Mask')

    cbar = fig.colorbar(im, ax=axs[0], orientation='vertical')
    cbar.set_label('Temperature')

    plt.show()

```
